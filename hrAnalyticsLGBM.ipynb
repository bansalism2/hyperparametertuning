```python
def encode_and_bind(original_dataframe, feature_to_encode):
    dummies = pd.get_dummies(original_dataframe[[feature_to_encode]])
    res = pd.concat([original_dataframe, dummies], axis=1)
    return(res)
```

```python
from lightgbm import LGBMClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn.model_selection import RandomizedSearchCV
```

```python
employee_train,employee_test=train_test_split(employee_data,test_size=0.3,random_state=142)
```

```python
train_y = employee_train['Attrition_Yes']
test_y = employee_test['Attrition_Yes']
```

```python
train_x = employee_train.drop(['Attrition_Yes'],axis=1)
test_x = employee_test.drop(['Attrition_Yes'],axis=1)
```

```python
lgbm_clf = LGBMClassifier(random_state=142)
```

```python
lgbm_clf.fit(train_x,train_y)
```

```python
y_pred = lgbm_clf.predict(test_x)
```

```python
#Random Search CV

n_estimators = [int(x) for x in np.linspace(start = 5, stop = 1000, num = 10)]
n_estimators.append(100)
learning_rate = [x for x in np.linspace(start=0.01, stop=0.3, num=5)]
learning_rate.append(0.1)
num_leaves = [int(x) for x in np.linspace(start = 5, stop = 1000, num = 10)]
num_leaves.append(31)
max_depth = [int(x) for x in np.linspace(start = 5, stop = 1000, num = 10)]
max_depth.append(-1)
min_child_samples = [int(x) for x in np.linspace(start = 0, stop = 200, num = 20)]
min_child_samples.append(20)
class_weight = ['balanced', None]
lgb_grid = {'n_estimators': n_estimators,
               'learning_rate': learning_rate,
               'num_leaves': num_leaves,
               'max_depth': max_depth,
                'min_child_samples':min_child_samples,
               'class_weight': class_weight}
```

```python
lgb_random_trials = pd.DataFrame(columns=['n_iter','cv','accuracy','time_taken','best_n_est',
                                        'best_lr','best_num_leaves','best_max_depth','best_min_child_samples',
                                        'best_class_weight'])
for n_iter in [5,10,20,50,100,200,500]:
    for cv in [2,3,5]:
        tic=time.time()
        lgb_random = RandomizedSearchCV(estimator = lgbm_clf, param_distributions = lgb_grid, n_iter = n_iter, cv = cv, verbose=-2, random_state=142, n_jobs = -1)
        # Fit the random search model
        lgb_random.fit(train_x, train_y)
        accuracy=lgb_random.best_score_
        best_params=lgb_random.best_params_
        best_n_est=best_params['n_estimators']
        best_lr=best_params['learning_rate']
        best_num_leaves=best_params['num_leaves']
        best_max_depth=best_params['max_depth']
        best_min_child_samples = best_params['min_child_samples']
        best_class_weight=best_params['class_weight']
#         best_bootstrap=best_params['bootstrap']
        time_taken=time.time()-tic
        lgb_random_trials = lgb_random_trials.append({'n_iter':n_iter,'cv':cv,'accuracy':accuracy,
                                                    'time_taken':time_taken,'best_n_est':best_n_est,
                                        'best_lr':best_lr,
                                                    'best_num_leaves':best_num_leaves,
                                                    'best_max_depth':best_max_depth,
                                                      'best_min_child_samples':best_min_child_samples,
                                        'best_class_weight':best_class_weight},ignore_index=True)
```

```python
#Bayesian Trials
from hyperopt import hp,fmin,tpe,STATUS_OK,Trials
```

```python
# !pip install hyperopt
```

```python
# Create the random grid

space = {'n_estimators' : hp.choice('n_estimators', n_estimators),
        'learning_rate': hp.choice('learning_rate', learning_rate),
        'num_leaves': hp.choice('num_leaves', num_leaves),
        'max_depth': hp.choice('max_depth', max_depth),
         'min_child_samples':hp.choice('min_child_samples',min_child_samples),
        'class_weight' : hp.choice('class_weight', class_weight)
    }
```

```python
from sklearn.model_selection import cross_val_score
```

```python
global cv_obj
def objective(space):
    model = LGBMClassifier(n_estimators = space['n_estimators'],
                                 learning_rate = space['learning_rate'],
                                 num_leaves = space['num_leaves'],
                                 max_depth = space['max_depth'],
                                 min_child_samples = space['min_child_samples'],
                                 class_weight = space['class_weight']
                                 )
    #5 times cross validation fives 5 accuracies=>mean of these accuracies will be considered
    accuracy = cross_val_score(model, train_x, train_y, cv = cv_obj).mean()
    # We aim to maximize accuracy, therefore we return it as a negative value
    return {'loss': -accuracy, 'status': STATUS_OK }
```

```python
lgb_bayesian_trials = pd.DataFrame(columns=['n_iter','cv','accuracy','time_taken','best_n_est',
                                        'best_lr','best_num_leaves','best_max_depth','best_min_child_samples',
                                        'best_class_weight'])
for n_iter in [5,10,20,50,100,200,500]:
    for cv in [2,3,5]:
        tic=time.time()
        cv_obj=cv
        trials = Trials()
        best = fmin(fn= objective,
                    space= space,
                    algo= tpe.suggest,
                    max_evals = n_iter,
                    trials= trials)
        tac = time.time()
        lgb_bayesian_trials = lgb_bayesian_trials.append({'n_iter':n_iter,'cv':cv,'accuracy':trials.best_trial['result']['loss']*-1,
                                                        'time_taken':tac-tic,'best_n_est':n_estimators[best['n_estimators']],
                                        'best_lr':learning_rate[best['learning_rate']],
                                                        'best_num_leaves':num_leaves[best['num_leaves']],
                                                        'best_max_depth':max_depth[best['max_depth']],
                                                          'best_min_child_samples':min_child_samples[best['min_child_samples']],
                                        'best_class_weight':class_weight[best['class_weight']]},ignore_index=True)
```

```python
global cv_obj
cv_obj=2

n_estimators_optuna = [int(x) for x in np.linspace(start = 5, stop = 1000, num = 10)]
n_estimators_optuna.append(100)
learning_rate_optuna = [x for x in np.linspace(start=0.01, stop=0.3, num=5)]
learning_rate_optuna.append(0.1)
num_leaves_optuna = [int(x) for x in np.linspace(start = 5, stop = 1000, num = 10)]
num_leaves_optuna.append(31)
max_depth_optuna = [int(x) for x in np.linspace(start = 5, stop = 1000, num = 10)]
max_depth_optuna.append(-1)
min_child_samples_optuna = [int(x) for x in np.linspace(start = 0, stop = 200, num = 20)]
min_child_samples_optuna.append(20)
class_weight_optuna = ['balanced', None]

def objective(trial):
    n_estimators = trial.suggest_categorical('n_estimators', n_estimators_optuna)
    learning_rate = trial.suggest_categorical('learning_rate', learning_rate_optuna)
    num_leaves = trial.suggest_categorical('num_leaves',num_leaves_optuna)
    max_depth = trial.suggest_categorical('max_depth',max_depth_optuna)
    min_child_samples = trial.suggest_categorical('min_child_samples',min_child_samples_optuna)
    class_weight = trial.suggest_categorical('class_weight',class_weight_optuna)
#     bootstrap = trial.suggest_categorical('bootstrap',b)
 

    clf = LGBMClassifier(
            n_estimators=n_estimators, learning_rate=learning_rate,
            num_leaves=num_leaves,max_depth=max_depth,
            min_child_samples=min_child_samples,
            class_weight=class_weight)
    return cross_val_score(
        clf,train_x,train_y, n_jobs=-1, cv=cv_obj).mean()
```

```python
#Optuna Trials - Without Pruning
lgb_optuna_trials = pd.DataFrame(columns=['n_iter','cv','accuracy','time_taken','best_n_est',
                                        'best_lr','best_num_leaves','best_max_depth','best_min_child_samples',
                                        'best_class_weight'])
for n_iter in [5,10,20,50,100,200,500]:
    for cv in [2,3,5]:
        tic=time.time()
        cv_obj=cv
        study = optuna.create_study(direction='maximize')
        study.optimize(objective, n_trials=n_iter)
        trial = study.best_trial
        print('Accuracy: {}'.format(trial.value))
        print("Best hyperparameters: {}".format(trial.params))
        tac = time.time()
        lgb_optuna_trials = lgb_optuna_trials.append({'n_iter':n_iter,'cv':cv,'accuracy':trial.value,
                                                        'time_taken':tac-tic,'best_n_est':trial.params['n_estimators'],
                                        'best_lr':trial.params['learning_rate'],
                                                        'best_num_leaves':trial.params['num_leaves'],
                                                        'best_max_depth':trial.params['max_depth'],
                                                        'best_min_child_samples':trial.params['min_child_samples'],
                                        'best_class_weight':trial.params['class_weight']},ignore_index=True)
```

```python
from sklearn_genetic import GASearchCV
from sklearn_genetic.space import Continuous, Categorical, Integer
from sklearn_genetic.plots import plot_fitness_evolution, plot_search_space
from sklearn.model_selection import StratifiedKFold
import matplotlib.pyplot as plt
```

```python
# param_grid = {'min_weight_fraction_leaf': Continuous(0.01, 0.5, distribution='log-uniform'),
#               'bootstrap': Categorical([True, False]),
#               'max_depth': Integer(2, 30),
#               'max_leaf_nodes': Integer(2, 35),
#               'n_estimators': Integer(100, 300)}

n_estimators = [int(x) for x in np.linspace(start = 5, stop = 1000, num = 10)]
n_estimators.append(100)
learning_rate = [x for x in np.linspace(start=0.01, stop=0.3, num=5)]
learning_rate.append(0.1)
num_leaves = [int(x) for x in np.linspace(start = 5, stop = 1000, num = 10)]
num_leaves.append(31)
max_depth = [int(x) for x in np.linspace(start = 5, stop = 1000, num = 10)]
max_depth.append(-1)
min_child_samples = [int(x) for x in np.linspace(start = 0, stop = 200, num = 20)]
min_child_samples.append(20)
class_weight = ['balanced', None]

param_grid = {'n_estimators': Categorical(n_estimators),
              'learning_rate': Categorical(learning_rate),
              'num_leaves':Categorical(num_leaves),
              'max_depth': Categorical(max_depth),
              'min_child_samples':Categorical(min_child_samples),
              'class_weight': Categorical(class_weight)}
```

```python
clf=LGBMClassifier(random_state=142)
```

```python
lgb_genetic_trials = pd.DataFrame(columns=['n_iter','cv','accuracy','time_taken','best_n_est',
                                        'best_lr','best_num_leaves','best_max_depth','best_min_child_samples',
                                        'best_class_weight'])
for n_iter in [1,2,5,10,20,50]:
    for cv_cnt in [2,3,5]:
        tic=time.time()
        cv_obj=cv_cnt
        cv = StratifiedKFold(n_splits=cv_obj, shuffle=True)

        evolved_estimator = GASearchCV(estimator=clf,
                               cv=cv,
                               scoring='accuracy',
                               population_size=10,
                               generations=n_iter,
                               tournament_size=3,
                               elitism=True,
                               crossover_probability=0.8,
                               mutation_probability=0.1,
                               param_grid=param_grid,
                               criteria='max',
                               algorithm='eaMuPlusLambda',
                               n_jobs=-1,
                               verbose=True,
                               keep_top_k=4)
        evolved_estimator.fit(train_x,train_y)
        tac = time.time()
        lgb_genetic_trials = lgb_genetic_trials.append({'n_iter':n_iter,'cv':cv_cnt,'accuracy':evolved_estimator.best_score_,
                                                        'time_taken':tac-tic,'best_n_est':evolved_estimator.best_params_['n_estimators'],
                                        'best_lr':evolved_estimator.best_params_['learning_rate'],
                                                        'best_num_leaves':evolved_estimator.best_params_['num_leaves'],
                                                        'best_max_depth':evolved_estimator.best_params_['max_depth'],
                                                        'best_min_child_samples':evolved_estimator.best_params_['min_child_samples'],
                                        'best_class_weight':evolved_estimator.best_params_['class_weight']},ignore_index=True)
```
