```python
def encode_and_bind(original_dataframe, feature_to_encode):
    dummies = pd.get_dummies(original_dataframe[[feature_to_encode]])
    res = pd.concat([original_dataframe, dummies], axis=1)
    return(res)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn.model_selection import RandomizedSearchCV
```

```python
employee_train,employee_test=train_test_split(employee_data,test_size=0.3,random_state=142)
```

```python
train_y = employee_train['Attrition_Yes']
test_y = employee_test['Attrition_Yes']
```

```python
train_x = employee_train.drop(['Attrition_Yes'],axis=1)
test_x = employee_test.drop(['Attrition_Yes'],axis=1)
```

```python
rfc = RandomForestClassifier(random_state=142)

rfc.fit(train_x,train_y)

y_pred = rfc.predict(test_x)

print('Accuracy: %.3f' % accuracy_score(test_y, y_pred))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(test_y,y_pred))
```

```python
#Random Search CV
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 1000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10,20, 50,100,200]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10,20, 50,100,200]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
```

```python
rf_random_trials = pd.DataFrame(columns=['n_iter','cv','accuracy','time_taken','best_n_estimators',
                                        'best_min_samples_split','best_min_samples_leaf','best_max_features',
                                        'best_max_depth','best_bootstrap'])
for n_iter in [5,10,20,50,100,200,500]:
    for cv in [2,3,5]:
        tic=time.time()
        rf_random = RandomizedSearchCV(estimator = rfc, param_distributions = random_grid, n_iter = n_iter, cv = cv, verbose=-2, random_state=42, n_jobs = -1)
        # Fit the random search model
        rf_random.fit(train_x, train_y)
        accuracy=rf_random.best_score_
        best_params=rf_random.best_params_
        best_n_estimators=best_params['n_estimators']
        best_min_samples_split=best_params['min_samples_split']
        best_min_samples_leaf=best_params['min_samples_leaf']
        best_max_features=best_params['max_features']
        best_max_depth=best_params['max_depth']
        best_bootstrap=best_params['bootstrap']
        time_taken=time.time()-tic
        rf_random_trials = rf_random_trials.append({'n_iter':n_iter,'cv':cv,'accuracy':accuracy,
                                                    'time_taken':time_taken,'best_n_estimators':best_n_estimators,
                                        'best_min_samples_split':best_min_samples_split,
                                                    'best_min_samples_leaf':best_min_samples_leaf,
                                                    'best_max_features':best_max_features,
                                        'best_max_depth':best_max_depth,'best_bootstrap':best_bootstrap},ignore_index=True)
```

```python
# Create the random grid

n_est = [int(x) for x in np.linspace(start = 200, stop = 1000, num = 10)]
m_d = [int(x) for x in np.linspace(10, 110, num = 11)]
m_d.append(None)
m_f = ['auto', 'sqrt',None]
m_s_l = [1, 2, 5, 10,20, 50,100,200]
m_s_s = [2, 5, 10,20, 50,100,200]
b = [True,False]

space = {'n_estimators' : hp.choice('n_estimators', n_est),
        'max_depth': hp.choice('max_depth', m_d),
        'max_features': hp.choice('max_features', m_f),
        'min_samples_leaf': hp.choice('min_samples_leaf', m_s_l),
        'min_samples_split' : hp.choice('min_samples_split', m_s_s),
         'bootstrap': hp.choice('bootstrap',b)   
    }
```

```python
global cv_obj
def objective(space):
    model = RandomForestClassifier(max_depth = space['max_depth'],
                                 max_features = space['max_features'],
                                 min_samples_leaf = space['min_samples_leaf'],
                                 min_samples_split = space['min_samples_split'],
                                 n_estimators = space['n_estimators'],
                                bootstrap = space['bootstrap']
                                 )
    #5 times cross validation fives 5 accuracies=>mean of these accuracies will be considered
    accuracy = cross_val_score(model, train_x, train_y, cv = cv_obj).mean()
    # We aim to maximize accuracy, therefore we return it as a negative value
    return {'loss': -accuracy, 'status': STATUS_OK }
```

```python
rf_bayesian_trials = pd.DataFrame(columns=['n_iter','cv','accuracy','time_taken','best_n_estimators',
                                        'best_min_samples_split','best_min_samples_leaf','best_max_features',
                                        'best_max_depth','best_bootstrap'])
for n_iter in [5,10,20,50,100,200,500]:
    for cv in [2,3,5]:
        tic=time.time()
        cv_obj=cv
        trials = Trials()
        best = fmin(fn= objective,
                    space= space,
                    algo= tpe.suggest,
                    max_evals = n_iter,
                    trials= trials)
        tac = time.time()
        rf_bayesian_trials = rf_bayesian_trials.append({'n_iter':n_iter,'cv':cv,'accuracy':trials.best_trial['result']['loss']*-1,
                                                        'time_taken':tac-tic,'best_n_estimators':n_est[best['n_estimators']],
                                        'best_min_samples_split':m_s_s[best['min_samples_split']],
                                                        'best_min_samples_leaf':m_s_l[best['min_samples_leaf']],
                                                        'best_max_features':m_f[best['max_features']],
                                        'best_max_depth':m_d[best['max_depth']],'best_bootstrap':b[best['bootstrap']]},ignore_index=True)
```

```python
n_est = [int(x) for x in np.linspace(start = 200, stop = 1000, num = 10)]
m_d = [int(x) for x in np.linspace(10, 110, num = 11)]
m_d.append(None)
m_f = ['auto', 'sqrt',None]
m_s_l = [1, 2, 5, 10,20, 50,100,200]
m_s_s = [2, 5, 10,20, 50,100,200]
b = [True,False]

global cv_obj
cv_obj=2

def objective(trial):
    n_estimators = trial.suggest_categorical('n_estimators', n_est)
    max_depth = trial.suggest_categorical('max_depth', m_d)
    max_features = trial.suggest_categorical('max_features',m_f)
    min_samples_leaf = trial.suggest_categorical('min_samples_leaf',m_s_l)
    min_samples_split = trial.suggest_categorical('min_samples_split',m_s_s)
    bootstrap = trial.suggest_categorical('bootstrap',b)
 

    clf = RandomForestClassifier(
            n_estimators=n_estimators, max_depth=max_depth,
            max_features=max_features,min_samples_leaf=min_samples_leaf,
            min_samples_split=min_samples_split,bootstrap=bootstrap)
    return cross_val_score(
        clf,train_x,train_y, n_jobs=-1, cv=cv_obj).mean()
```

```python
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=2)
trial = study.best_trial
print('Accuracy: {}'.format(trial.value))
print("Best hyperparameters: {}".format(trial.params))
```

```python
#Optuna Trials - Without Pruning
rf_optuna_trials = pd.DataFrame(columns=['n_iter','cv','accuracy','time_taken','best_n_estimators',
                                        'best_min_samples_split','best_min_samples_leaf','best_max_features',
                                        'best_max_depth','best_bootstrap'])
for n_iter in [5,10,20,50,100,200,500]:
    for cv in [2,3,5]:
        tic=time.time()
        cv_obj=cv
        study = optuna.create_study(direction='maximize')
        study.optimize(objective, n_trials=n_iter)
        trial = study.best_trial
        print('Accuracy: {}'.format(trial.value))
        print("Best hyperparameters: {}".format(trial.params))
        tac = time.time()
        rf_optuna_trials = rf_optuna_trials.append({'n_iter':n_iter,'cv':cv,'accuracy':trial.value,
                                                        'time_taken':tac-tic,'best_n_estimators':trial.params['n_estimators'],
                                        'best_min_samples_split':trial.params['min_samples_split'],
                                                        'best_min_samples_leaf':trial.params['min_samples_leaf'],
                                                        'best_max_features':trial.params['max_features'],
                                        'best_max_depth':trial.params['max_depth'],'best_bootstrap':trial.params['bootstrap']},ignore_index=True)
```
